```python
# import libraries and read the data
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_sales = pd.read_csv('sales.csv')
```

```python
# data overview
print(df_sales.shape)
print(df_sales.info())
```

```python
# check missing values
# seems like it is fine to drop the salesdate missing values
print(df_sales.isnull().sum())
```

```python
# data feature engineering

## change column names to lowercase
df_sales.columns = df_sales.columns.str.lower()
print(df_sales.columns)

## salesdate column feature engineering
df_sales['salesdate'] = pd.to_datetime(df_sales['salesdate'])
df_sales['year'] = df_sales['salesdate'].dt.year
df_sales['month'] = df_sales['salesdate'].dt.month
print(df_sales.head(2))
```

```python
# check time span: only 5 month

## year
max_year = df_sales['year'].max()
min_year = df_sales['year'].min()
print(max_year, min_year)

## month
max_month = df_sales['month'].max()
min_month = df_sales['month'].min()
print(max_month, min_month)
```
